```python
from tensorflow import keras
import numpy as np
```

```python
model= keras.Sequential([keras.layers.Dense(units=1,input_shape=[1])])
#the optimizer is 'stochastic gradient descent' and the loss function is MSE
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
#Declare the x and y vectors as numpy arrays y = f(x)= 3x+2
x = np.array([-3.0,-2.0, -1.0, 0.0 , 1.0, 2.0, 3.0], dtype=float)
y = np.array([-7.0,-4.0, -1.0, 2.0, 5.0, 8.0, 11.0], dtype=float)
```

```python
model.fit(x,y, epochs=500)
print("The predicted output value is ", model.predict([10.0]))
```

After 500 epochs the model finds the relationship between the x and y inputs ( in this case it's linear as y= 3x+2) , for an input value of x= 10 , f(10) = 32
The model predicts 31.99998 which is close enough to 32
